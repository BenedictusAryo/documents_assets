Import package pandas 

```python
import pandas as pd
```

Inputkan data dari http://bit.ly/drinksbycountry. Data berikut merupakan data konsumsi minuman dengan satuan  litres per capita of drinking-age population.

```python
df = pd.read_csv("http://bit.ly/drinksbycountry")
df.head()
```

Data dari informasi dari tabel diatas mempunyai 6 variabel yaitu country (liter), beer servings (liter), spirit servings (liter), wine servings (liter), total litres of pure alcohol dan continent. 

Melihat info data

```python
df.info()
```

Dari info data diatas terdapat baris data sebanyak 193 baris data. Dengan 193 baris data untuk semua variabel baik country, beer servings, spirit servings, wine servings, total litres of pure alcohol dan continent. Oleh karena itu, diartikan bahwa nilai pada semua variabel terisi atau tidak ada data yang kosong. 

Melihat banyaknya negara setiap benua (continent) 

```python
df['continent'].value_counts()
```

Dari data diatas menyampaikan banyak baris data (banyaknya negara) sesuai dengan Benua (continent) yaitu Afrika sebanyak 53 negara, Europe sebanyak 45 negara, Asia sebanyak 44 negara dan seterusnya dapat dilihat pada output diatas. 

Melihat deskripsi statistika data untuk semua data

```python
df.describe()
```

Dari tabel diatas menghitung analisa statistika deskriptif. Pada beer servings mempunyai rata-rata 106.16 liter dari semua negara, Standar deviasi sebesar 101.143 liter, Nilai minimal atau nilai terkecil sebesar 0 liter, Quartil 1 (25%) sebesar 20 liter, Quartil 2 (50%) sebesar 76 liter, Quartil 3 (75%) sebesar 188 liter dan nilai maximal atau nilai tertinggi yaitu sebesar 376 liter, dan seterusnnya untuk variabel lain dapat dilihat pada tabel output diatas.  

Mengurutkan data sesuai dengan variabel Benua (continent) dan urutkan sesuai abjad, tujuannya agar melihat data sesuai dengan kelompok Benua dengan mudah. 

```python
df1 = df.sort_values('continent').reset_index(drop=True)
df1.head()
```

Misalnya, kita ingin mengetahui data dari negara Indonesia 

```python
df[df['country']=="Indonesia"]
```

Diketahui bahwa data dari negara Indonesia dengan beer servings sebesar 5 liter, spirit servings sebesar 1 liter, wine servings sebesar 0 liter dan total litres of pure alcohol sebesar 0.1 liter. Disimpulkan dari rata-rata pernegara negera Indonesia mempunyai minuman yang sangat kecil. 

Total semua konsumsi dari semua negara pada data set.

```python
df.sum(numeric_only=True)
```

beer servings mepunyai total konsumsi dari semua negara untuk beer servings sebesar 20489 liter, spirit servings sebesar 15632 liter, wine servings sebesar 9544 dan total litres of pure alcohol sebesar 910 liter. 

Total konsumsi minuman berdasarkan Benua (continent)

```python
df.groupby('continent', as_index=False).sum()
```

Pada benua Afrika total konsumsi minuman beer servings sebesar 3158 liter, Benua Asia 1630 liter, Benua Eropa 8720  liter, dan seterusnya dapat dilihat pada tabel diatas. Didapatkan konsumsi beer survings paling tinggi yaitu Benua Eropa dengan 8720 liter dan Benua dengan konsumsi terendah Oceania dengan 1435 liter. Dan lainnya dapat dilihat pada tabel diatas. 

Melihat nilai konsumsi tertinggi pada setiap benua. 

```python
df4 = df.groupby(['continent'], as_index=False).max()
df4.drop(columns='country')
```

Dari tabel diatas konsumsi tertinggi pada Benua Afrika untuk minuman beer servings sebesar 376 liter, sirit serving sebesar 152 liter, wine serving sebesar 233 liter dan total litres of pure alcohol sebesar 9.1 liter. Dan Benua lainnya dapat dilihat pada tabel diatas. 

Rata-rata konsumsi minuman setiap negara berdasarkan benua. 

```python
df.groupby(['continent'], as_index=False).mean()
```

Rata-rata konsumsi minuman beer serving pada benua Afrika yaitu sebesar 61.47 liter, spirit serving dengan rata-rata 16.34 liter, wine serving dengan rata-rata 16.26 liter dan total litres of pure alcohol sebesar 3.01 liter. 

Menampilkan konsumsi terendah dari setiap benua

```python
df5 = df.groupby(['continent'], as_index=False).min()
df5.drop(columns='country')
```

Pada benua afrika konsumsi terendah semua minuman yaitu 0 liter. Benua lainnya dapat dilihat pada tabel diatas. 

Membandingkan konsumsi minuman indonesia dan Malaysia

```python
df[(df['country']=="Indonesia") | (df['country']=="Malaysia")]
```

syntax diatas dapat digunakan lagi dengan cara berikut

```python
negara = ['Indonesia', 'Malaysia']
df[df['country'].isin(negara)]
```

Melihat data dibenua Asia

```python
df2 = df[df['continent'] == 'Asia'].reset_index(drop=True)
df2
```

Menampilkan data pada benua Asia dengan variabel country dan beer saja.
