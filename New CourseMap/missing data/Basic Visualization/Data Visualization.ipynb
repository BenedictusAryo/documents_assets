# Import your data

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/athlete_events.csv')
df.head()
```

# Deal with Missing Values

```python
df['Year'].unique()
```

```python
df.shape
```

```python
df = df[df['Year']==2016]
df.shape
```

```python
df = df.fillna({'Medal':'None'})
```

```python
df.head()
```

```python
df.info()
```

# How to Change Categorical into Numerical

```python
df.head()
```

```python
all_go = pd.DataFrame(df.groupby(['NOC','Medal'])['Medal'].size())
```

```python
all_go.head()
```

```python
all_go.columns = ['Total']
```

```python
all_go.head()
```

```python
all_go = all_go.reset_index()
all_go.head()
```

```python
all_go['None'] = np.where(all_go['Medal']=='None', all_go['Total'],0)
all_go['G']= np.where(all_go['Medal'] == 'Gold', all_go['Total'], 0)
all_go['S']= np.where(all_go['Medal'] == 'Silver', all_go['Total'], 0)
all_go['B']= np.where(all_go['Medal'] == 'Bronze', all_go['Total'], 0)
```

```python
all_go.head()
```

```python
all_go = all_go.groupby('NOC').sum().reset_index().sort_values(by=['G','S','B'],ascending=False)
```

```python
all_go['Total_Medals'] = all_go['Total'] - all_go['None']
```

```python
all_go.head()
```

# Bar Chart

```python
plt.figure(figsize=(12,12))
df['Sport'].value_counts().plot(kind='barh')
plt.title('Total Athlete on Sport in Olympic 2016')
plt.xlabel('Total Athlete')
plt.ylabel('Sports')
plt.show()
```

# Line Chart

```python
df1 = pd.read_csv('data/athlete_events.csv')
df1.head()
```

```python
all_new = pd.DataFrame(df1.groupby(['Year','NOC','Medal'])['Medal'].size())
```

```python
all_new.head()
```

```python
all_new.columns = ['Total']
```

```python
all_new.head()
```

```python
all_new = all_new.reset_index()
```

```python
all_new.head()
```

```python
all_new['None'] = np.where(all_new['Medal']=='None', all_new['Total'],0)
all_new['G']= np.where(all_new['Medal'] == 'Gold', all_new['Total'], 0)
all_new['S']= np.where(all_new['Medal'] == 'Silver', all_new['Total'], 0)
all_new['B']= np.where(all_new['Medal'] == 'Bronze', all_new['Total'], 0)

all_new = all_new.groupby(['Year','NOC']).sum().reset_index().sort_values(by=['G','S','B'],ascending=False)
```

```python
all_new.head()
```

```python
gre = all_new[all_new['NOC']=='GRE']
```

```python
gre
```

```python
sns.lineplot(x='Year',y='Total',data=gre)
```
