# Pemilihan tabel, dan penyingkatan data 

Panggil package pandas 

```python
import pandas as pd
```

Inputkan data data_penjualan_motor2018.xlsx yang disimpan pada file data dan menampilkan 5 data pertama dengan menggunakan .head(). 

```python
motor = pd.read_excel('data/data_penjualan_motor2018.xlsx')
print(motor.head())
```

Menjumlahkan variabel yang bertipe numerik dalam data penjualan motor 2018 ini yaitu variabel jumlah. 

```python
motor.sum(numeric_only=True)
```

Dari output diatas dapat diketahui jumlah penjualan motor pada tahun 2018 yaitu sebanyak 6382780 unit motor baik dari motor jenis Honda, Yamaha, Kawasaki maupun Suzuki. 

Melihat jumlah penjualan sesuai dengan Bulan.

```python
motor.groupby('Bulan', as_index=False).sum()
```

Dari tabel diatas dappat diketahui bahwa jumlah penjualan pada  bulan Agustus sebanyak 567961 unit, Bulan April 580917, dan seterusnya dapat dilihat pada tabel diatas. Perlu diketahui Bulan yang akan mengurutkan sesuai abjad. 

Melihat Jumlah penjualan dengan urutan variabel Bulan kemudian variabel Jenis. 

```python
motor.groupby(['Bulan','Jenis'], as_index=False).sum()
```

Dari tabel diatas pada Bulan Agustus, Jenis Honda terjual sebanyak 443694 unit, Jenis Kawasaki sebanyak 7016 unit, Jenis Suzuki sebanyak 7016 unit dan Yama sebanyak 108896 Unit. Data selanjutnya dapat dilihat pada tabel diatas.

Melihat rata-rata penjualan

```python
motor.groupby(['Jenis'], as_index=False).mean()
```

Rata-rata penjualan Jenis Honda sebesar 396600 unit, Kawasaki sebesar 6582 unit, Suzuki sebanyak 7459 unit dan Yamaha sebanyak 121257. 

Melihat pengurutan Jumlah motor dari besar ke kecil dengan Bulan dan Jenis. 

```python
totals = motor.sort_values('Jumlah', 
                           ascending=False).reset_index(drop=True)
totals
```

Dari tabel diatas didapatkan jumlah penjualan tertinggi yaitu pada bulan April dengan Jenis Honda dengan 458499 unit, tertinggi kedua yaitu pada bulan Oktober dengan Jenis Honda berjumlah 4586582 unit, dan seterusnya dalat dilihat pada tabel diatas.

Membuat function untuk melihat Maximal, mean dan minimal data. 

```python
def get_stats(group):
    return {'min': group.min(), 'max': group.max(), 'mean': group.mean()}
conclusion = motor['Jumlah'].groupby(motor['Jenis']).apply(get_stats).unstack()
conclusion
```

Pada tabel diatas, pada Jenis motor Honda penjualan terbanyak dengan 458499 unit, dengan rata-rata penjualan 396600 unit dan penjualan terkecil 271206 unit, dan seterusnya bisa dilihat pada tabel diatas.
